```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
```

Apply vader sentiment analyser to some sentences

```python
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
    print("Overall sentiment dictionary is : ", sentiment_dict)
    print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        print("Positive")
 
    elif sentiment_dict['compound'] <= - 0.05 :
        print("Negative")
 
    else :
        print("Neutral")
   
# Driver code
if __name__ == "__main__" :
 
    print("\n1st statement :")
    sentence = "that sports game was awful"
 
    # function calling
    sentiment_scores(sentence)
 
    print("\n2nd Statement :")
    sentence = "the late late show was so bad tonight i hate the presenter "
    sentiment_scores(sentence)
 
    print("\n3rd Statement :")
    sentence = "what a dreadful show that was."
    sentiment_scores(sentence)
    
    print("\n4th Statement :")
    sentence = "that was great"
    sentiment_scores(sentence)
    
    
    print("\n5th Statement :")
    sentence = "i dont know whether to be happy or sad"
    sentiment_scores(sentence)
    
    print("\n6th Statement :")
    sentence = "load of rubbush"
    sentiment_scores(sentence)
    
    print("\n7th Statement :")
    sentence = "i dont understand the person talking"
    sentiment_scores(sentence)
    
 
```

Read in annotated tweets

```python
tweets_corpus = pd.read_csv("AnnotatedTweets.csv", index_col=0)
```

```python
tweets_corpus.head()
```

```python
analyzer = SentimentIntensityAnalyzer()
tweets_corpus['compound'] = [analyzer.polarity_scores(x)['compound'] for x in tweets_corpus['text']]
tweets_corpus['neg'] = [analyzer.polarity_scores(x)['neg'] for x in tweets_corpus['text']]
tweets_corpus['neu'] = [analyzer.polarity_scores(x)['neu'] for x in tweets_corpus['text']]
tweets_corpus['pos'] = [analyzer.polarity_scores(x)['pos'] for x in tweets_corpus['text']]
```

```python
tweets_corpus
```

```python
def sentiment(label):
    if label < 0:
        return "Negative"
    elif label == 0:
        return "Neutral"
    elif label > 0:
        return "Positive"
```

```python
tweets_corpus["sentiment"] = tweets_corpus["compound"].apply(sentiment)
```
